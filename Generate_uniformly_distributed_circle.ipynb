# How to generate a uniformly distributed circle?

```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt
%matplotlib inline


n = 2000
rho = [np.sqrt(random.randint(0,n)) for x in range(n)]
theta = [random.randint(0, n)*(2*math.pi/n) for x in range(n)]
x = []
y = []
for i in range(n):
    x.append(rho[i] * math.cos(theta[i]))
    y.append(rho[i] * math.sin(theta[i]))
        
fig, ax = plt.subplots(nrows=1, ncols=1, figsize = (10,10))
ax = plt.subplot(111)
s = ax.scatter(x, y, s=57,marker=("o"), cmap=plt.cm.hsv, linewidths=0, alpha=0.7)
```
